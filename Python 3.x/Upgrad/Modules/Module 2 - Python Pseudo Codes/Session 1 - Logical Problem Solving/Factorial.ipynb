## Factorial of a number

Another simple application of Pseudo Codes is the factorial. We can easily calculate the factorial of any number by using some simple lines of code. In this case, we follow the following logic  



There can be 2 methods to execute this  

__Using `while` loop__

Since we do not know the number of iterations in the loop, we would use `while` loop, and run it such that it would keep multiplying the variable `fact` with the iterable variable `i` until it equals `n`.

We would also like to see if the value entered is 0 or not. If it is 0, then fact will be 1 only. Let us see the code for this 

```python
n=int(input('Enter the number:'))
i,fact=1,1;
while i<=n:
    fact=fact*i if n>0 else 1;
    i+=1;
    
print('{}! = {}'.format(n,fact)) if n>=0 else print('error!')
```

__Using the `reduce() lambda` function__

This way, we will first take input `n` from the user, create a list from 1 to the number `n`, and then use the `reduce()` function to multiply all the consecutive elements in the list. We also want to put conditions in such a manner that for `n=0`, we have `n!=1`. 

Let us see the code for this -

```python
from functools import reduce
n=int(input('Enter the number:'))

fact=reduce(lambda x,y:x*y,[i for i in range(1,n+1)]) if n>0 else 1;
print('{}! = {}'.format(n,fact)) if n>=0 else print('error!')
```

This is how the factorial of a number is calculated.
