## Reversing the Digits of an Integer

Another straightforward application of pseudocode is to reverse the digits of an integer. In this case, we will first take input `n`, convert it into string, reverse it, and then convert the reversed string into number. 

If the input value is negative, we will first multiply it by -1 to remove the - sign, then use the above process to convert and then multiply the reversed integer with -1. Let us see the flow diagram here- 



This flow diagram describes the algorithm perfectly. Now, let us see the code used to apply this logic - 

```python
n=int(input('Enter the number:'))

rev=int(str(n*-1)[-1:(len(str(n))*-1)-1:-1])*-1 if n<0 else int(str(n)[-1:(len(str(n))*-1)-1:-1])

print(rev)
```

Hence, it is just a one-line code used to implement such a long algorithm. This is the efficiency of python !
