### II) Superstore Data

In the `superstore.csv` file, you have the details of orders purchased in an American online retail store. Load the dataset, observe and analyse the different columns and answer the following questions.

```python
#Load the dataset
df2 =pd.read_csv('https://github.com/yashj1301/Python3-UpGrad-UMich/raw/master/Python%203.x/Upgrad/Modules/Module%204%20-%20Data%20Visualization/Data/superstore.csv')
df2.head()
```

 
**Q1)** Write the code to plot only those data points for which the Sales were less than 15,000 and made a loss.

```python
df3=df2[(df2.Sales<15000)&(df2.Profit<0)]
sns.jointplot(x='Sales',y='Profit',data=df3)
```

**Q2)** Ship Mode column has 4 categories – First Class, Second Class, Same Day and Standard Class, while the Segment has 3 categories – Consumer, Home Office and Corporate.

Which of the following can be used to visualise how the average Profit varies across every Segment - Ship Mode combination (like Consumer - First Class, Corporate - Standard Class and so on)?

- Histogram
- Box Plot
- Scatter Plot
- Heatmap


```python
sns.heatmap(df2.pivot_table(index='Segment',columns='Ship Mode', values='Profit', aggfunc=np.mean),cmap='Reds',annot=True)
plt.show()
```

**Q3)** Write the code that would help you create a bar plot in seaborn that compares the average Sales across the different Segment types?

```python
sns.barplot(x='Segment',y='Sales', data=df2, estimator=np.mean)
```

**Q4)** Plot a pie-chart to find the Ship Mode through which most of the orders are being delivered.

     a)Standard Class
     b)First Class
     c)Second Class
     d)Same Day
     

```python
#Your code here
df2['Ship Mode'].value_counts().plot(kind='pie',autopct='%.3g',labels=None)
plt.ylabel('Ship Mode ', rotation=0)
plt.legend(df2['Ship Mode'].value_counts().index,loc=(0.7,0.6))
plt.show()
```
