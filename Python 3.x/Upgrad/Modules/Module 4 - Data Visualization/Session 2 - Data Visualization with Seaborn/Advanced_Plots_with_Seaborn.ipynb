## Advanced Plots

There are many plots that have somewhat complicated functioning in the PyPlot module of the Matplotlib library. However, using Seaborn, we can create them without much fuss. These plots are â€“
- Regression Plots
- Pair Plots
- Heatmaps
- Stacked Bar Charts and Box Plots
- Line Charts

Let us learn about them one by one. First, let us import all the necessary libraries. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as cb
import warnings as warn

warn.filterwarnings('ignore')
```

Now that we have imported all the libraries, it is time to get going. We will start with the regression plots. These plots can easily plot regression models.


### Regression Plots

Seaborn provides __automatic estimation and plotting for regression setups__ for different kinds of variables. So, let us get an intuitive understanding of what insights they provide us with. We use the `jointplot()` method of seaborn, with the parameter `kind` where we specify it as `reg` (short for regression). Let us have a look at it now.

```python
x=np.random.randint(0,100,100)
y=np.random.randint(23,189,100)
```

```python
cb.jointplot(x,y,kind='reg')
```

A __Regplot__ is nothing but a regression line added to a jointplot (along with a KDE curve over the histogram). This line tries to generalise the relationship between the two numeric variables in the form of a linear regression equation, in simpler terms, it finds a straight line passing through the data that also best describes the data.

Next, let us learn about pairplots. These are used to create a grid of scatter plots for all the numeric variables in a dataframe.

### Pair Plots

Now, in case there are 4â€“5 numeric variables that you want to analyse, making a jointplot for every 2 numeric variables is a bit tedious. To overcome this limitation, we use the pair plots or pairplots (used interchangeably throughout). It is done using the `pairplot()` method of seaborn.

These plots __create scatter plots for each numeric variable in a dataframe against all the other numeric variables__ in the dataframe. It takes the <font color="red">dataframe as argument, alongside the keyword argument `kind`</font>, which offers the functionality of regression line fitting. 

It returns a grid of __<code>n<sup>2</sup></code>__ elements, where __`n`__ is the number of numeric variables in the dataframe. It follows the following syntax -

__`cb.pairplot(dataframe,kind= )`__

Let us see an example.

```python
z=np.random.randint(123,130,100)

new=pd.DataFrame({'x':[i for i in x],'y':[i for i in y],'z':[i for i in z]})

cb.pairplot(new)
plot.show()
```

As you can see, the pair plot creates a scatter-plot for every two numeric variables, whereas in the case of the diagonal ones where the same variable is being considered twice, a histogram is shown. Here, you can make certain inferences about the mathematical relationships between these variables. You can also see that the grid returned is of <code>3<sup>2</sup>=9</code> elements. 

Next, let us create Bar Graphs and Box Plots using Seaborn.

### Bar Graphs and Box Plots

We learnt how bar graphs and box plots can be utilised for analysing the numerical and categorical variables in your data. Now, we will learn about some additional customisations that Seaborn provides, along with certain use cases where those functionalities come in handy.

```python
rate=pd.read_csv('https://github.com/yashj1301/Python3-UpGrad-UMich/blob/master/Python%203.x/Upgrad/Modules/Module%204%20-%20Data%20Visualization/Data/ratings.csv?raw=true',
                 parse_dates=True,index_col=0)
rate.head()
```

```python
rate.info()
```

This is our dataset. First, let us change the reviews and minimum installs columns data types to integers. 

```python
rate.Reviews=rate.Reviews.astype(int)
rate['Min. Installs']=rate['Min. Installs'].astype(int)
rate.info()
```

Now, let us create our barplots. We will be using the `barplot()` method of seaborn, which was previously understood. 

```python
cb.barplot(data=rate, x='Content Rating', y='Rating')
plot.xticks(rotation=90)
plot.tight_layout()
```

If we want, we can change the functionalities inside the barplot, by defining the aggregation we want it to display. This can be done using the __`estimator`__ argument of this method. It takes either an inbuilt NumPy function, or a custom lambda function as value. 

Let us see it in action.

```python
cb.barplot(data=rate, x='Content Rating', y='Rating', estimator=np.min)
plot.xticks(rotation=90)
plot.tight_layout()
```

We can see that we have plotted the ratings of each category by their minimum values, and the variation against the minimum value (shown by the vertical lines). This is how we can extend the functionality of bar plots. 

However, while dealing with numeric data, such plots are not sufficient for us to understand the data. Hence, we use boxplots for such scenarios. 

This is done using the __`cb.boxplot()`__ method, that takes either one or two arguments. If one argument is given, then the plotting of only a single variable is shown, but if there are two arguments, then the plotting of the variables against each other is shown. It follows the given syntax -

__`cb.boxplot(x,y)`__

Let us see it in action.

```python
plot.subplots(figsize=(10,7))
cb.boxplot(rate['Content Rating'],rate['Rating'])
plot.tight_layout()
```

We can clearly see how the boxplot is plotted for the `Content Rating` and `Rating` columns against each other. What if we want to create a boxplot for just some manually selected categories ?

We will select the top 4 categories based on `Genres` and plot them against their ratings. Let us see.

```python
top4=list(rate['Genres'].value_counts().sort_values(ascending=False)[:4].index)
plot.subplots(figsize=(10,7))
cb.boxplot(rate[rate['Genres'].isin(top4)]['Genres'],rate['Rating'])
plot.show()
```

This is how we can create boxplots to get valuable insights about our data. These kind of boxplots are really helpful in finding relationships between categorical and numeric variables.

### Heatmaps

If you have used MS Excel, then you must have come across conditional formatting, which utilises colour intensities to indicate the magnitude of the numerical value like the one below:

image.png

Heat maps use a similar principle of visualising a range of values by utilising colours and colour intensities. They convert numerical columns into categorical columns using binning, and assign colors to each entry on the basis of those bins. This is how a heatmap is created.

For using heatmaps, let us first import a dataset. 

```python
new=pd.read_csv('https://github.com/yashj1301/Python3-UpGrad-UMich/blob/master/Python%203.x/Upgrad/Modules/Module%204%20-%20Data%20Visualization/Data/ratings.csv?raw=true',
                parse_dates=True,index_col=0)
new.head()
```

The dataset is imported. For heatmaps, we first have to create categorical columns relative to our numerical data. Let us see some categorical columns we already have. 

```python
new['Genres'].value_counts()
```

For our analysis, let us take the top 5 categories from here. The following code will be used for this. 

```python
res=new[new.Genres.isin(new.Genres.value_counts()[:5].index)]
res.Genres.value_counts()
```

Now, it is time to create our heatmap. We will be plotting it alongside our other categorical variable - `Content Rating`. It has the following categories -

```python
res['Content Rating'].value_counts()
```

Now, we will create the heatmap. For this purpose, we use the __`heatmap()`__ method available in `seaborn`. This method takes __a dataframe as an argument, that has both rows and columns as categorical data__, and returns a heatmap of that. It also takes 2 keyword arguments - `cmap` (color map) and `annot` (asking if text should be visible or not). It follows the following syntax -

__`cb.heatmap(dataframe,cmap= ,annot= )`__

Let us see it in action. For heatmaps, we often use the `pivot_table()` functionalities of Pandas. Let us create the heatmap now.

```python
res_heatmap=res.pivot_table(index='Genres',columns='Content Rating',values='Rating',aggfunc=lambda x:np.quantile(x,0.1)).fillna(res['Rating'].mode()[0])
plot.subplots(figsize=(7,5))
cb.heatmap(res_heatmap,cmap='Spectral',annot=True)
plot.show()
```

The heatmap is created. The color scale shows the range of values, while the labels on each box is depicting the values of 10th percentile for each corresponding category. To make it more readable, we can change the cmap color to a more unicolor format. Let us see. 

```python
plot.subplots(figsize=(7,5))
cb.heatmap(res_heatmap,cmap='Reds',annot=True)
plot.show()
```

See? A single color scale has made it much more readable. Now, we can also create a new categorical variable in the dataframe, and create a pivot table for heatmap. 

For this, pandas has a method called the __`qcut()`__ method, which takes 3 parameters, on the basis of percentiles - 

- `Series` / `DataFrame` on the basis of whose values, the cateogries need to be created.
- `List of quantiles` (percentiles) in increasing order, that need to be created for binning.
- `List of Categorical Values` for each bin, in increasing order. 

It follows the following syntax -

__`pd.qcut(series/dataframe,[list of quantiles],[list of bin values])`__

Let us see it in action.

```python
new['Bucket']=pd.qcut(new['Size'],[0,0.2,0.4,0.6,0.8,1],['VL','L','M','H','VH'])
new.head()
```

Now, to create the heatmap, we will plot the `Bucket` column against the `Content Rating` column on the basis of 20th percentile values of `Rating` for each entry. Let us create the heatmap data. 

```python
heatmap_data=new.pivot_table(index='Content Rating',columns='Bucket',
                             values='Rating', aggfunc=lambda x:np.quantile(x,0.2)).fillna(new['Rating'].mean()).round(4)

heatmap_data
```

Now, the only that is left, is to create a heatmap for the given data. Let us use the `heatmap()` method to create it. 

```python
plot.subplots(figsize=(7,5))
cb.heatmap(heatmap_data,cmap='Greens',annot=True)
plot.show()
```

This is how we create heatmaps. We can even modify some other keyword arguments for our heatmap. The `fmt` argument is used for string formatting for decimals. Keenly notice the difference in data labels of the heatmap above and the one below. 

```python
plot.subplots(figsize=(7,5))
cb.heatmap(heatmap_data,cmap='Oranges',annot=True, fmt='.3g')
plot.show()
```

These functionalities of Seaborn make it a crucial asset to any data scientist. Next, we will see how to create stacked bar charts.

### Stacked Bar Charts

We have already learnt how to create bar charts using both `matplotlib`, and `seaborn` libraries. However, we havenâ€™t created __stacked bar charts__ â€“ where we use another categorical variable in the mix to create a multivariate graph. We will do that now. 

In seaborn, we use the __`hue`__ argument inside the __`barplot()`__ method of seaborn to create multivariate stacked bar charts. Let us import a dataset first to create these charts.


```python
rate=pd.read_csv('https://github.com/yashj1301/Python3-UpGrad-UMich/blob/master/Python%203.x/Upgrad/Modules/Module%204%20-%20Data%20Visualization/Data/ratings.csv?raw=true',
                 parse_dates=True,index_col=0)
rate.info()
```

We will be using the `Content Rating` vs the `Rating` graph, where we will see how many of each category are paid and free, based on the column `Type`. The following pivot table will explain this.

```python
pivot=rate.pivot_table(index='Content Rating',columns='Type', values='Rating',aggfunc=lambda x:np.quantile(x,0.1)) 
pivot
```

We can clearly see that the categories `Unrated` and `Adults only 18+` have very insignificant numbers relative to others, so we won't consider them. Let us do just that.

```python
pivot.drop(pivot[pivot.Paid.isnull()].index,axis=0,inplace=True)
pivot
```

Now, let us create our stacked bar chart. We will use the __`plot()`__ method here, with the `kind` argument as `Bar` and `stacked` argument set to <code><font color="#000080">True</font></code>. Let us see it.

```python
cb.set_style('darkgrid')
pivot.plot(kind='bar',stacked=True,color=['green','red'], title='Stacked Bar Chart')
plot.ylabel('10th percentile of No. of Reviews')
plot.legend(loc=(1.02,0.8))
plot.show()
```

We can also create a horizontal stacked chart, using the `kind` argument in the `plot()` method as `barh`. Let us see this.

```python
cb.set_style('darkgrid')
pivot.plot(kind='barh',stacked=True,color=['green','red'], title='Stacked Bar Chart')
plot.ylabel('10th percentile of No. of Reviews')
plot.legend(loc=(1.02,0.8))
plot.show()
```

We can also create stacked percentage bar charts. Let us see. We can also use the `colormap` argument to set a color scheme for the plot. Let us see the following example. 

```python
filt=pivot.apply(lambda x:x/x.sum(),axis=1)
filt.plot(kind='bar',stacked=True,colormap='tab20b',title='Stacked Bar Chart')
plot.ylabel('10th percentile of No. of Reviews')
plot.legend(loc=(1.02,0.8))
plot.show()
```

We cannot create stacked charts using Seaborn, but we can create multivariate bar graphs. This is done using the `hue` parameter of the __`barplot()`__ method. Let us see.

```python
new=rate[rate['Content Rating'].isin(filt.index)]
plot.subplots(figsize=(7,5))
plot.title('Multi-Variate Bar Chart using Seaborn')
cb.barplot(x='Rating',y='Content Rating',data=new,hue='Type',
           estimator=lambda x:np.quantile(x,0.2),palette='Reds',orient='horizontal')
plot.legend(loc=(1.02,0.8))
plot.show()
```

This is how we can create multivariate bar charts and stacked bar charts using matplotlib and seaborn.
