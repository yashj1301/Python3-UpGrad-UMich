## Data Visualization Using Seaborn

This library is extensively used in python to __create aesthetically beautiful, sophisticated__ graphs. It is also used to __further extend the functionality__ of our existing plots –
- Line Graphs and Scatter Plots
- Bar Charts and Histograms
- Box Plots

This library is made from matplotlib and is closely integrated with Pandas. It is a dataset-oriented API that offers the following functionalities – 
- It can be used alongside the `PyPlot` module of matplotlib for better visualizations
- It has options of exploring both univariate and bivariate data
- It automatically estimates and plots linear regression models
- It offers convenient views for complex datasets
- It gives you control over your color palette and styling of plots

These features of Seaborn make it an essential tool in the toolkit of a data scientist. Now, to use this library, we use the `import` statement, using the alias `cb`. It follows the given syntax –

__`import seaborn as cb`__

Now, we are ready to use this library. Let us import all the necessary libraries to our notebook first.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as cb
```

All the libraries are imported. Now, it is time to learn how to use this library. We will start with Distribution Plots.

### Distribution Plots

First, we will plot a numpy array using the `pyplot` module, and afterwards, we will move towards the `seaborn` plotting. 

Python Seaborn module contains various functions to plot the data and depict the data variations. The __`seaborn.distplot()`__ method is used to plot the distplot. The distplot represents the univariate distribution of data i.e. data distribution of a variable against the density distribution. It follows the following syntax -

__`seaborn.distplot(data,bins,color)`__

Let us see.

```python
x=np.random.randint(1,5,100)
plot.hist(x,bins=4)
plot.grid()
```

Now, let us plot the same array using the `seaborn` library. For this, let us use the `distplot()` method. Let us see it in action.

```python
cb.distplot(x, bins=4, color='red')
plot.grid()
```

We can clearly see 3 things -
- First, the distribution line is drawn upon the bar charts. This line is called KDE (Probability Density Curve).
- Second, it is showing the probability distribution chart, instead of our normal histogram. 
- Third, we can use the `matplotlib` methods alongside `seaborn` methods for our visualizations.

The seaborn histogram also takes an additional parameter, `kde`, which when set to `False` won't show our Probability Distribution line. Let us see it in action.

```python
fig,ax = plot.subplots(1,2, figsize=(8,4), sharey=False)

plot.subplot(1,2,1)
plot.title('Normal Histogram')
plot.ylabel('Frequency')
plot.xlabel('Range')
cb.distplot(x,bins=3,color='green',kde=False)
plot.grid()

plot.subplot(1,2,2)
plot.title('KDE Histogram')
cb.distplot(x,bins=3,color='darkblue',kde=True)
fig.tight_layout(h_pad=10)
plot.grid()

plot.show()
```

See? This is how we use seaborn to create amazing visualizations for our plots. Let us see some customizations here.

### Styling Options from Seaborn

One of the __major advantages of adopting Seaborn__, as previously mentioned, is that <font color='darkblue'>you may keep its aesthetic qualities while simultaneously using Matplotlib to perform extra customisations</font>. Let’s look at some of the styling options available in Seaborn.

#### Styling with the `set_style()` Method

Out of the many ways to customize a plot, the easiest one to use is the `set_style()` method available in seaborn. 

This method is used to define a style palette for your plot, without writing much codes. It includes various in-built themes that have defined colors, styles and functionalities for various chart elements, and just by mentioning the style name, you can access the whole theme. 

It takes two keyword arguments - 
- `style`, which defines either the in-built style name, or a custom style created using dictionaries (similar to the `fontdict` parameter in pyplot), and
- `rc`, which takes a dictionary as argument, that overrides the default values for the style. 

Syntax - __`cb.set_style( style= , rc= )`__

Let us see it in action.

```python
cb.set_style('dark')
plot.title('KDE Histogram')
cb.distplot(x,bins=3,kde=True)
plot.show()
```

Here, we can see that the distribution plot returned is aesthetically much more pleasing. We can even change the parameters inside this style, using the `rc` argument. Let us do that.

```python
fig,ax = plot.subplots(2,2, sharey=False, figsize=(12,6))

cb.set_style('dark')
plot.subplot(2,2,1)
plot.title('Dark Theme')
cb.distplot(x,bins=3,kde=True, color='yellow')

cb.set_style('white')
plot.subplot(2,2,2)
plot.title('White Theme')
cb.distplot(x,bins=3,kde=True, color='brown')

cb.set_style('darkgrid')
plot.subplot(2,2,3)
plot.title('DarkGrid Theme')
cb.distplot(x,bins=3,kde=True, color='violet')

cb.set_style('whitegrid')
plot.subplot(2,2,4)
plot.title('WhiteGrid Theme')
cb.distplot(x,bins=3,kde=True, color='yellowgreen')

fig.tight_layout()
plot.show()
```

There are many more styles available in the Seaborn library. Apart from this, we also have styles available in the `style` method of the `pyplot` module of `Matplotlib` library. Let us explore more of them. 

This method `style` is used before the plotting functions to apply the styling. It has a method called `use()` in which we have to specify the name of the style we want to use. Let us now see it in action.

```python
fig,ax = plot.subplots(figsize=(6,4))
plot.style.use('dark_background')
plot.hist(x,bins=3,color='green')
plot.show()
```

When we use the `plot.style.use()` method, we <font color='red'>change the default values</font> of our plot styles. Now, even if we won't use the above statement, even then we will get the last used style, which has been set. Let us see an example.

```python
cb.distplot(x,bins=4,color='white')
```

Here, you can see that the `dark_background` theme is set, even though we did not use the plot.style.use() method. To reset it back, we will use the same command, and pass the value `default` as argument. This will reset all the modifications in the style settings we did. 

Let us do this now.

```python
plot.style.use("default")
```

```python
cb.distplot(x,bins=4)
plot.show()
```

Hence, we can see that the styling has been reset. This is how we can use styling in plots for creating better visualizations. We also see some warnings coming before our plots. 

These are just some backend warnings that need not be considered. To hide them, we use the `filterwarnings()` method from the library `warnings`. This will help us in hiding these warnings everytime we create our charts. Let us see.

```python
import warnings as warn
warn.filterwarnings('ignore')
```

```python
plot.subplots(figsize=(4,3))
cb.distplot(x,bins=4)
plot.show()
```

See? No more warnings, and the presentation of our notebook also becomes much better. This is how we use `seaborn` for some basic visualizations.
