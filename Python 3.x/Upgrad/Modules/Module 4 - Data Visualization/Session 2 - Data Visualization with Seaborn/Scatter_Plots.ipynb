## Scatter Plots

Previously, you had dealt with a single numeric column and, therefore, used either a box plot or a histogram to portray the insights visually. What about two numeric columns? 

If you want to plot the relationship between two numeric variables, you will be using something known as a <b>Scatter Plot</b>. 

Let us import all the necessary libraries first.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as cb
import warnings as warn

warn.filterwarnings('ignore')
```

Now that we have imported all the necessary libraries, it is time to get going. We have already used a scatterplot using the __`plot.scatter()`__ method of the __`matplotlib`__ library. Let us do that first. 

```python
x=np.random.randint(0,100,100)
y=np.random.randint(0,100,100)

plot.title('Scatter Plot using the pyplot Module')
plot.xlabel('x')
plot.ylabel('y')
plot.scatter(x,y,marker='.',color='green')
plot.grid()
```

This is how the scatter plot is created using the PyPlot module. Scatterplots are one of the most widely used and powerful visuals in the field of machine learning. They are very important for identifying associations between data points, and you can usually derive some form of trend in the data using a scatterplot.

### Some Applications of Scatterplot in Machine Learning

There are certain use cases where a scatterplot is immensely productive, especially in the field of machine learning:

- <b><u>Observing trends in linear regression</u></b>: Because scatterplots can reveal patterns in the data, they’re a necessity in linear regression problems where you want to determine whether making a linear model, i.e. using a straight line to predict something makes sense or not.
- <b><u>Observing natural clusters in the data</u></b>: In simple terms, clustering is the act of grouping similar entities into clusters. For example, let’s say you have a group of students who have recently taken a test in maths and biology.You can clearly divide the students into four clusters - Cluster 1 is made up of children who excel in biology but struggle in maths, Cluster 2 is made up of individuals who excel in both courses, and so on.

You already know how to do this in matplotlib using the `pyplot.scatter()` function. In seaborn, we have the `sns.scatterplot()`, which is pretty intuitive and similar to its matplotlib counterpart.

### Scatter Plots using Seaborn

We have seen how we can make use of the `PyPlot` module of the `matplotlib` library to create scatter plots. In seaborn also, we have similar functionality. 

Seaborn has the seaborn.scatterplot() method to create scatterplots, in much better aesthetics than matplotlib, while providing greater flexibility in customization. It takes the following parameters -

- `x` and `y` - The numerical variables that needed to be plot.
- `data` - The dataframe from which x and y are columns.
- `hue` - The grouping variable.
- `size` - This parameter allows us to display different categories data with different colors.
- `palette` - This parameter allows us to change the color palette of the chart.
- `size_norm`
- `size_order`

It follows the following syntax for creating scatterplots - 

__`cb.scatterplot(x,y,data= ,hue= ,size= ,palette= ,size_norm= ,size_order= )`__

Let us see it in action.

```python
x=np.random.randint(0,100,100)
y=np.random.randint(0,100,100)

cb.set_style('darkgrid')
plot.title('Scatter Plot using Seaborn')
plot.xlabel('x')
plot.ylabel('y',rotation='horizontal')
cb.scatterplot(x,y,palette='deep',color='green')
plot.show()
```

We can see how different, and elegant our scatter plot becomes. Let us now depict how we can create a scatterplot with categories.

For this, we will create a dataframe having columns of x, y and cat (True or False). Next, we will create a scatterplot of x and y, grouping on the cat column. We will achieve this by using the `random.choice()` method of NumPy in list comprehension, which allots a random value out of a list of string values to each element of a list. Let us see it in action.

```python
x=np.random.randint(0,100,100)
y=np.random.randint(123,231,100)
cat=[np.random.choice([True,False]) for i in x]

fig,ax = plot.subplots(2,2,figsize=(12,8))

plot.subplot(2,2,1)
df=pd.DataFrame({'x':x,'y':y,'Bool':cat})
cb.scatterplot(x='x',y='y',hue='Bool',data=df,palette='gnuplot')
plot.ylabel('y\n',rotation='horizontal')

plot.subplot(2,2,2)
df=pd.DataFrame({'x':x,'y':y,'Bool':cat})
cb.scatterplot(x='x',y='y',hue='Bool',data=df,palette='bone')
plot.ylabel('y\n',rotation='horizontal')

plot.subplot(2,2,3)
df=pd.DataFrame({'x':x,'y':y,'Bool':cat})
cb.scatterplot(x='x',y='y',hue='Bool',data=df,palette='Pastel1')
plot.ylabel('y\n',rotation='horizontal')

plot.subplot(2,2,4)
df=pd.DataFrame({'x':x,'y':y,'Bool':cat})
cb.scatterplot(x='x',y='y',hue='Bool',data=df,palette='gist_earth')
plot.ylabel('y\n',rotation='horizontal')

plot.tight_layout()
plot.show()
```

We can clearly see how convenient it is to create multi-variate scatter plots using seaborn. Now, we will learn about another type of scatter plot in python - Joint plot. These plots are used to display a histogram and a scatterplot together.

### Joint Plots

Joint Plots are used to draw a plot of two variables with bivariate and univariate graphs. This is achieved through the jointplot() method in seaborn.

This method provides a convenient interface to the ‘JointGrid’ class, with several canned plot kinds. It offers the following parameters -

- `x` and `y` - hese parameters take Data or names of variables in the parameter `data`.
- `data` - This parameter take DataFrame when `x` and `y` are variable names.
- `kind` - This parameter take Kind of plot to draw.
- `color` - This parameter take Color used for the plot elements.
- `dropna` - This parameter take boolean value, If True, remove observations that are missing from `x` and `y`.

It follows the following syntax to create jointplots -

__`cb.jointplot(x,y,data= ,kind= ,color= ,ratio= ,space= ,dropna= )`__

Let us now see it in action.

```python
cb.set_style('darkgrid')
cb.jointplot(x='x',y='y', data=df,color='darkred',dropna=True)
plot.tight_layout()
plot.show()
```

In addition to the normal scatterplot, __the jointplot adds the histogram of the respective columns to the mix__. This way, you can get an idea of the spread of the variables being discussed and make more succinct conclusions and gather insights from the data.
