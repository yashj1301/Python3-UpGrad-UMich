## Bar Graphs

Bar graphs are the pictorial representation of data (generally grouped), in the form of vertical or horizontal rectangular bars, where the length of bars is proportional to the measure of data.

They are very useful in understanding the dimensional data types. They can be used for studying categories, classifying data, and finding similar insights. 

To use bar charts, we use the `bar()` method, that has the following syntax - 

`plot.bar(data,height)`

Let us see it in action.

```python
import numpy as np
import matplotlib.pyplot as plot
```

```python
x=np.linspace(1,6,6,dtype=int)
y=np.array([3,4,6,2,4,2])
x,y
```

```python
plot.title('\nBar Chart for Frequency\n')
plot.xlabel('Index')
plot.ylabel('Frequency')
plot.bar(x,y)
plot.grid(axis='y',color="#d3d3d3")
```

### Enhancing Bar Graphs

For enhancing bar graphs, we can simply use the `fontdict` and the `color` parameters to give it a new look. The color parameter can have a list of colors as values, which will repeat the colors in the same sequence, in case the number of bars is greater than the list of colors. 

Let us see them in action.

```python
title,label={'family': 'serif',
        'color': 'blue',
        'weight': 'normal',
        'size': 18,
        },{'family': 'serif',
        'color': 'blue',
        'weight': 'normal',
        'size': 14,
        }
```

```python
plot.title('\nBar Chart for Frequency\n',fontdict=title)
plot.xlabel('Index',fontdict=label)
plot.ylabel('Frequency',fontdict=label)
plot.bar(x,y,color=['red','blue','green'])
plot.grid(axis='y',color="#d3d3d3")
```

See? Our graph is looking much better than before. Now, let us try to rename the tick values (values displaying on the axis). We can do that by using the `xticks()` and `yticks()` methods. 

These methods take the ___lists of original values and the new values__ as arguments. The `xticks()` method will modify for the `x-axis`, and `yticks()` will modify for `y-axis`. Let us see them in action. 

```python
x_labels=['One','Two','Three','Four','Five','Six']
y_labels=['{}M'.format(i) if i!=0 else i for i in [0,1,2,3,4,5,6]]
```

```python
plot.title('\nBar Chart for Frequency\n',fontdict=title)
plot.xlabel('Index',fontdict=label)
plot.ylabel('Frequency',fontdict=label)
plot.bar(x,y,color=['red','blue','green'])
plot.xticks(x,x_labels)
plot.yticks([0,1,2,3,4,5,6],y_labels)
plot.grid(axis='y',color="#d3d3d3")
```

See? This is how we can use bar graphs for visualization.

### Horizontal Bar Charts

We can easily create horizontal bar charts as well. Let us look at them. 

```python
plot.title('\nBar Chart for Frequency\n',fontdict=title)
plot.xlabel('Frequency',fontdict=label)
plot.ylabel('Category',fontdict=label)
plot.barh(x,y,color=['red','blue','green'])
plot.xticks([0,1,2,3,4,5,6],y_labels)
plot.yticks(x,x_labels)
plot.grid(axis='x',color="#d3d3d3")
```
