## Scatter Plots

Scatterplot, as the name suggests, displays the __scatter of the data__. It can be helpful in checking for any __relationship pattern__ between two __quantitative variables__ and __detecting the presence of outliers__ within them.

```python
import numpy as np
import matplotlib.pyplot as plot
```

To create a scatter plot, we use the `scatter()` method of the pyplot module. This method takes 2 required arguments for each axis, and other than that, it takes some keyword arguments like color and label. 

Syntax - `plot.scatter(x,y,color=,label=)`

We can also use all the other chart element methods like `grid()`, `xlabel()`, `ylabel()`, `xticks()`, `yticks()` and `title()`. Let us see it in action now. 

```python
x=np.random.randint(35,234,50)
y=np.random.randint(30,200,50)

x,y
```

```python
plot.title('Scatter Plot of X versus Y')
plot.xlabel('X-values')
plot.ylabel('Y-values')
plot.scatter(x,y,color='brown')
plot.grid('y')
```

This plot showcases the scattering of data points across the two arrays. We can clearly see that there doesn't seem to be much relationship between the two variables. This is where the data comes in handy. 

Let us now create multi-variable scatter plot that depicts relationships.

```python
a1=np.arange(0,2*np.pi+0.01,np.pi/9)
plot.title('\nGraph of sin(x) and cos(x) for [0,2pi]\n',fontdict={'family':'bahnschrift','size':18})
plot.xlabel('x in radians',fontdict={'family':'consolas','size':14})
plot.ylabel('sin(x) and cos(x) Values',fontdict={'family':'consolas','size':14})
plot.scatter(a1,np.sin(a1),color='red')
plot.scatter(a1,np.cos(a1),color='green')

plot.legend(['y=sin(x)','y=cos(x)'])
plot.grid()
```

This scatter plot can clearly tell us that there is a fourier transform relationshop between the two plotted variables. Hence, it is important for us to understand how scatter plot helps us in finding relationships. 

There is another method, called `annotate()` which helps us put labels on certain data points inside our graph. This method takes two keyword arguments - 

- <font face="consolas">text</font> - This parameter defines the string that needs to be displayed. 
- <font face="consolas">xy</font> - This parameter defines the tuple of coordinates where the data point is located. 

Let us say, in the above graph, we want to put a label on all those points where `cos(x)=1`. For this, we use the `annotate()` function. We will use a for loop to extract all those points that have cos(x)=1 and pass them into the keyword argument `xy`.

Let us see how it works.

```python
a1=np.arange(-np.pi/9,2*np.pi+1,np.pi/9)
plot.title('\nGraph of sin(x) and cos(x) for [0,2\u03C0]\n',fontdict={'family':'bahnschrift','size':18})
plot.xlabel('x in radians',fontdict={'family':'consolas','size':14})
plot.ylabel('sin(x) and cos(x) Values',fontdict={'family':'consolas','size':14})
plot.scatter(a1,np.sin(a1),color='red')
plot.scatter(a1,np.cos(a1),color='green')

# using the annotate function here

for coords in zip(a1,np.cos(a1)):
    if coords[1]==1: plot.annotate(text='cos(n\u03C0)',xy=coords)

plot.legend(['sin(x)','cos(x)'])
plot.grid()
```

As you can see, the data points where <font face="cambria math">cos(x)=1</font> are labelled using the `annotate()` method. Hence, this is how we create and use `scatter-plots` for visualization.
