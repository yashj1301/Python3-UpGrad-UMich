## Introduction to Matplotlib

We have learnt how to handle data using the NumPy library. You could load data in the form of arrays and then perform operations on the same. Now, we will try to visualise the arrays using another library in Python, namely, __`matplotlib`__.

__Data Visualization__ is an essential component of data science. It is used for understanding data, finding out underlying patterns, relationships and outliers, and most important – it is used to communicate your findings to the stakeholders.

The <code>__matplotlib__</code> Library in python, is used for creating powerful visualizations of datasets. Combining it with the NumPy library, we have a crucial data toolkit used by data scientists worldwide. 

Before learning about visualization, let us learn about the types of data – facts and dimensions below. These are essential in understanding our data.

### Facts and Dimensions

Graphics and visuals, if used intelligently and innovatively, can convey a lot more than raw data alone. Matplotlib serves the purpose of providing multiple functions to build graphs from the data stored in your lists, arrays, etc.

Before we start discussing different plots, we need to understand the elements that will help us prepare charts and plots effectively. Data are of two types, namely:

- <u>Facts / Measures</u> – This data contains numeric or quantitative values that can be measured. It represents facts, and you can apply calculations and aggregations on them.
- <u>Dimensions</u> – This data contains qualitative information, such as names, dates, or location information. This data can be used to classify, categorize, group, or segment your data. 

Before plotting a chart, it is important to understand the type of data you have – this would lead to various visualization techniques that are specific to the data type. Facts are numerical data, and dimensions are the metadata – data explaining other data – associated with the fact variables. Both facts and dimensions are equally important for generating actionable insights from a data set.

### Visualization Plots

`Plots` are used to convey different ideas. For example, using some plots, you may want to visualise the spread of data across two variables. With some others, you may want to gauge the frequency of a certain label. Depending on the objective of your visualisation task, you will choose an appropriate plot.

Before doing that, we first __need to know how to use the Matplotlib library__. It uses the same syntax as the NumPy library, only that __our alias is <font face="consolas">plot__</font>. 

Let us see the syntax now. We will be using the `pyplot` module of the library for visualization.

```python
import matplotlib.pyplot as plot
```

Now that we have accesses the library, it is time to get to know more about the `pyplot` module. We will use the `help()` function to learn about all the features of this library.

```python
help(plot)
```

### Visualization using `Pyplot` Module

To create charts, we have to use some pre-defined methods in `PyPlot`, that are used to create visualizations.

In general, whenever we use the `plot` alias, we will generate a plot object, which does not get shown automatically in some cases. Hence, we use the following method to show them - 

- <font face="consolas">plot.show()</font> - This method is used to send the output chart to the console.

For each type of graph, we have different methods to create them. Let us first start with the line graph.
Let us now see them in action.


#### Line Graph

Line Graphs are used to visualize data that has been studied over a period of time, or data that is showing some relationship between 2 variables x and y. 

To create a line graph, we use the following method:  
- `plot.plot(x-axis, y-axis)` - This method is used to define the variables for the x-axis and y-axis of the visualization chart. 

Now, let us see it in action. 

```python
import numpy as np
```

Now, first we will create 2 arrays. First one will be our index - from 0 to 20, and the other one will be our random numerical data.

```python
x,y=np.arange(1,21),x*2.5
x,y
```

Now, let us plot this data. Our `plot()` method will automatically show <font face="consolas">line graph</font> for it. Let us see this now. 

```python
plot.plot(x,y)
plot.show()
```

There are some additional parameters for the `plot()` method that we can add for making our plot look good. These are - 

- <font face="consolas">markers</font>: This parameter lets us define various styles of markers on the graph. Some common styles are given below - 

|Style|Description|
|-----|-----------|
|'o'|Circle|
|'s'|Square|
|'D'|Diamond|
|'.'|Point|
|'*'|Star|
|','|Pixel|

- <font face="consolas">linestyle</font> or <font face="consolas">ls</font>: This parameter is only for line graphs. It lets us style the lines displayed on the chart. We have options of `dashed`, `dotted`, `solid` (default) and `dashdot`.

- <font face="consolas">color</font> or <font face="consolas">c</font>: This parameter allows us to change the color of the line displayed.
- <font face="consolas">width</font> or <font face="consolas">lw</font>: This parameter lets us define the width of the line displayed on the chart. 

Let us use these now.

```python
plot.plot(x,y,marker='D',ls='dashed',color="brown",lw=2)
```

See? This is how we can use various parameters to create a plot. 

#### Plotting Multiple Variables

We can also __plot multiple variables__ on the same plot. We can use the indexes of arrays `x` and `y` as x-axis, and their values as y-axis. Let us see this in action now.

```python
plot.plot(x,color="green")
plot.plot(y,color="red")
plot.show()
```

#### Methods for Chart Elements

Now, let us use some additional methods to insert plot elements like title, labels, and grid. We will add chart title, axes labels, grids and other enhancements here. We use the following methods for this - 

- <font face="consolas">xlabel()</font> and <font face="consolas">ylabel()</font> - These methods is used to display labels for the x-axis and the y-axis respectively. 
- <font face="consolas">title()</font> - This method is used to name the title of the chart. 
- <font face="consolas">grid(axis)</font> - This method is used to show gridlines in the chart. It has a parameter `axis` which lets us define the axis on which should the grid be displayed. 
- <font face="consolas">legend([list])</font> - This method is used to show the legends of the chart. It takes a list as argument that contains the label name for each legend in sequence x,y,z... till end.

Let us see them in action - 

```python
plot.title('Using indices as x-axis and values as y-axis\n')
plot.xlabel('\nindex')
plot.ylabel('values\n')
plot.plot(x,color="green")
plot.plot(y,color="red")
plot.legend(['x','y'])
plot.grid()
```

We also have a parameter __`fontdict`__, which is used to modify the font specifications inside the graph. For this, we create multiple dictionaries of font, and then specify those dictionaries inside the methods. It can be used on the `title()` method, or on the `xlabel()` and `ylabel()` methods, or even on the `legend()` method. 

Let us see it in action.

```python
title_font= {'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 16,
        }

label_font={'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 14,
        }
```

```python
plot.title('\nUsing Indices as X-axis and values as Y-axis\n',fontdict=title_font)
plot.xlabel('\nIndex',fontdict=label_font)
plot.ylabel('Values\n',fontdict=label_font)
plot.plot(x,color="green")
plot.plot(y,color="red")
plot.legend(['x','y'])
plot.grid(axis='y')
```

Hence, this is how we use the `matplotlib pyplot` package to create visualizations.
