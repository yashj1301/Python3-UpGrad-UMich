## Sub Plots

Sometimes, it is beneficial to draw different plots on a single grid next to each other to get a better overall view. Different plots presented in a single plot object are commonly referred to as __Subplots__. 

Let's learn how to create subplots inside a single plot in Matplotlib. To create subplots, we use two methods of the `pyplot` module – 

- <font face="consolas">plot.subplots(rows,cols,sharey,sharex)</font> – This method is used to define the template of the chart grid. It contains the following arguments –
    - number of rows, 
    - number of columns
    - <font face="consolas">sharex</font> - whether the x-axis should be shared among subplots or not
    - <font face="consolas">sharey</font> – whether the y-axis should be shared among the subplots or not. This method returns two objects – a map object, and an axes object.
<br><br>
- <font face="consolas">plot.subplot(row<sub>n</sub>,col<sub>n</sub>,loc)</font> – This method is used to write the block of code within a particular subplot. The code between two subsequent statements of this method will define a subplot code block. It has the following arguments - 
    - the number of rows of the grid
    - the number of columns of the grid
    - the location of the subplot

The location of the subplot is found out by numbering the subplots in increasing order. For example, if the grid is of 3x3, then it will have the following locations for the subplots –

|Column 1|Column 2|Column 3|
|------------------|----------------|-----------------|
|Subplot 1 (3,3,1)|Subplot 2 (3,3,2)|Subplot 3 (3,3,3)|
|Subplot 4 (3,3,4)|Subplot 5 (3,3,5)|Subplot 6 (3,3,6)|
|Subplot 7 (3,3,7)|Subplot 8 (3,3,8)|Subplot 9 (3,3,9)|

It is important to note that <font color='darkred'>if you call the <font face="consolas">show()</font> method inside any subplot, then the plots after it would be treated as separate to the current grid</font>. Hence, it is important to __call the <font face="consolas">show()</font> method only after the last subplot__ code block. 

Let us see it in action.

```python
import numpy as np
import matplotlib.pyplot as plot
```

```python
x=np.arange(0,2*np.pi,0.1)
```

```python
plot.subplots(2,2,sharey=True)

plot.subplot(2,2,1)

plot.title('Line Graph of sin(x)')
plot.xlabel('x (in radians)')
plot.ylabel('sin(x)')
plot.plot(x,np.sin(x),color='darkred')
plot.grid()

plot.subplot(2,2,2)

plot.title('Scatter Graph of sin(x)')
plot.xlabel('x (in radians)')
plot.scatter(x,np.sin(x),s=3,color='green')
plot.grid()

plot.subplot(2,2,3)
plot.title('Line Graph of cos(x)')
plot.xlabel('x (in radians)')
plot.ylabel('cos(x)')
plot.plot(x,np.cos(x),color='darkred')
plot.grid()

plot.subplot(2,2,4)

plot.title('Scatter Graph of cos(x)')
plot.xlabel('x (in radians)')
plot.scatter(x,np.cos(x),s=3,color='green')
plot.grid()

plot.show()
```

Here, we can see that the __labels and titles of the graphs are overlapping__. To address this issue, we use a method in `pyplot` called the `tight_layout()` method, which has a keyword argument `h_pad` (short for horizontal padding).

Before that, we mentioned above that the `subplots()` method returns two objects – map object and axes object. This `tight_layout()` method is to be used with the map object, and hence, we first have to create identifiers for these two objects. 

After this, we will use the `tight_layout()` method to the identifier `fig`, in which we will define the horizontal padding. Now, we can use the previous code to get our charts in proper format.

Let us see the final result now.

```python
fig,ax=plot.subplots(2,2,sharey=True)
fig.tight_layout(h_pad=4)

plot.subplot(2,2,1)

plot.title('Line Graph of sin(x)',fontdict={'family':'consolas','size':14})
plot.xlabel('x (in radians)',fontdict={'family':'consolas','size':12})
plot.ylabel('sin(x)',fontdict={'family':'consolas','size':12})
plot.plot(x,np.sin(x),color='darkred')
plot.grid()

plot.subplot(2,2,2)

plot.title('Scatter Graph of sin(x)',fontdict={'family':'consolas','size':14})
plot.xlabel('x (in radians)',fontdict={'family':'consolas','size':12})
plot.scatter(x,np.sin(x),s=3,color='green')
plot.grid()

plot.subplot(2,2,3)
plot.title('Line Graph of cos(x)',fontdict={'family':'consolas','size':14})
plot.xlabel('x (in radians)',fontdict={'family':'consolas','size':12})
plot.ylabel('cos(x)',fontdict={'family':'consolas','size':12})
plot.plot(x,np.cos(x),color='darkred')
plot.grid()

plot.subplot(2,2,4)

plot.title('Scatter Graph of cos(x)',fontdict={'family':'consolas','size':14})
plot.xlabel('x (in radians)',fontdict={'family':'consolas','size':12})
plot.scatter(x,np.cos(x),s=3,color='green')
plot.grid()

#plot.show()
```
